```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('form.csv')
```

```python
df
```

```python
def replace(cell):
    if (cell == 'Totalmente en desacuerdo'):
        return 1
    elif (cell == 'En desacuerdo'):
        return 2
    elif (cell == 'Ni en acuerdo ni en desacuerdo'):
           return 3
    elif (cell == 'De acuerdo'):
        return 4
    elif (cell == 'Totalmente de acuerdo'):
        return 5
    else:
        return cell
```

```python
df_processed = df.applymap(replace)
df_processed
```

```python
for i, text in enumerate(df.columns.values):
    print(f'{i} - {text}')
```

## Evaluacion Programa

```python
functionality = df_processed.iloc[:,:15]
functionality
```

```python
usability = pd.concat([df_processed.iloc[:,:2], df_processed.iloc[:,15:25]], axis='columns')
usability
```

```python
utility = pd.concat([df_processed.iloc[:,:2], df_processed.iloc[:,25:34]], axis='columns')
utility
```

```python
other_app = pd.concat([df_processed.iloc[:,:2], df_processed.iloc[:,34:36]], axis='columns')
other_app
```

```python
utility_manual = pd.concat([df_processed.iloc[:,:2], df_processed.iloc[:,36:39]], axis='columns')
utility_manual
```

```python
other_manual = pd.concat([df_processed.iloc[:,:2], df_processed.iloc[:,39:]], axis='columns')
other_manual
```

```python
# Write excel
with pd.ExcelWriter('output.xlsx') as writer:  
    functionality.to_excel(writer, sheet_name='functionality')
    usability.to_excel(writer, sheet_name='usability')
    utility.to_excel(writer, sheet_name='utility')
    other_app.to_excel(writer, sheet_name='other_app')
    utility_manual.to_excel(writer, sheet_name='utility_manual')
    other_manual.to_excel(writer, sheet_name='other_manual')
```
